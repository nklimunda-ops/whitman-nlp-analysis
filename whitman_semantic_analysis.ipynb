# Detecting Semantic Fields and Bodily Metonymy in Walt Whitman with Python and spaCy

This project analyzes a passage from Walt Whitman’s *Song of Myself* using Python and spaCy.

The goal is to identify semantic fields related to life, death, the body, nature, and spirituality, while exploring bodily metonymy and metaphorical patterns in the poem.

This project combines computational linguistics with literary analysis and digital humanities methodologies.

```python
import spacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")
```

```python
texto = """
A child said What is the grass? fetching it to me with full hands;
How could I answer the child? I do not know what it is any more than he.
I guess it must be the flag of my disposition, out of hopeful green stuff woven.

Or I guess it is the handkerchief of the Lord,
A scented gift and remembrancer designedly dropt,
Bearing the owner's name someway in the corners, that we may see and remark, and say Whose?

Or I guess the grass is itself a child, the produced babe of the vegetation.

Or I guess it is a uniform hieroglyphic,
And it means, Sprouting alike in broad zones and narrow zones,
Growing among black folks as among white,
Kanuck, Tuckahoe, Congressman, Cuff, I give them the same, I receive them the same.

And now it seems to me the beautiful uncut hair of graves.

Tenderly will I use you curling grass,
It may be you transpire from the breasts of young men,
It may be you are from old people, or from offspring taken,
It may be if I had known them I would have loved them, soon out of their mothers' laps,
And here you are the mothers' laps.

This grass is very dark to be from the white heads of old mothers,
Darker than the colorless beards of old men,
Dark to come from under the faint red roofs of mouths.

O I perceive after all so many uttering tongues,
And I perceive they do not come from the roofs of mouths for nothing.

I wish I could translate the hints about the dead young men and women,
And the hints about old men and mothers, and the offspring taken soon out of their laps.
What do you think has become of the young and old men?
And what do you think has become of the women and children?

They are alive and well somewhere,
The smallest sprout shows there is really no death,
And if ever there was it led forward life, and does not wait at the end to arrest it,
And ceas'd the moment life appear'd.

All goes onward and outward, nothing collapses,
And to die is different from what any one supposed, and luckier.
"""

doc = nlp(texto)
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_)
```

```python
semantic_fields = {
    "life": ["life", "alive", "child", "sprout", "vegetation", "offspring", "babe"],
    "death": ["death", "dead", "die", "grave"],
    "body": ["hand", "breast", "lap", "head", "beard", "mouth", "tongue", "hair"],
    "nature": ["grass", "green", "sprout", "vegetation", "earth"],
    "spirituality": ["lord", "soul", "heaven"]
}
```

```python
results = {}

for field, words in semantic_fields.items():
    count = 0
    found_words = []

    for token in doc:
        lemma = token.lemma_.lower()

        if lemma in words:
            count += 1
            found_words.append(token.text)

    results[field] = {
        "count": count,
        "words": found_words
    }

results
```

```python
for field, data in results.items():
    print(field.upper())
    print("Count:", data["count"])
    print("Words:", data["words"])
    print()
```

```python
metonymy_body_parts = ["hand", "breast", "lap", "head", "beard", "mouth", "tongue", "hair"]

metonymy_results = []

for token in doc:
    if token.lemma_.lower() in metonymy_body_parts:
        metonymy_results.append((token.text, token.lemma_, token.pos_))

metonymy_results
```

The poem uses body parts such as hands, breasts, laps, heads, beards, mouths, tongues, and hair.
These body parts can be read as metonymic images because they represent human presence, age, maternity, voice, death, and collective identity.

```python
grass_sentences = []

for sent in doc.sents:
    if "grass" in sent.text.lower():
        grass_sentences.append(sent.text)

for sent in grass_sentences:
    print(sent)
    print("---")
```

Grass is presented through a series of shifting metaphors: flag, handkerchief, child, hieroglyphic, and hair of graves.
This suggests that grass functions as a symbolic bridge between nature, body, death, and rebirth.

This analysis shows that Whitman’s passage connects nature, body, life, and death through repeated semantic patterns.
The computational analysis identifies a strong presence of bodily imagery and life/death vocabulary.
The body-related words are especially important because they function as possible metonymies: hands, mouths, tongues, breasts, and laps do not merely describe body parts, but represent people, generations, voice, maternity, and mortality.
The poem’s central image, grass, becomes a metaphorical object that links democratic equality, spiritual mystery, death, and renewal.
